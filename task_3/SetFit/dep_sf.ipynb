```python
from setfit import SetFitModel, Trainer, TrainingArguments
from datasets import Dataset
import pandas as pd
from sklearn.metrics import classification_report
```

```python
train_df = pd.read_csv('your/train/data.csv')
test_df = pd.read_csv('your/test/data.csv')

train_df['label'] = train_df['predicted_department']
test_df['text'] = test_df['position']
test_df['label'] = test_df['department']
```

```python

train_ds = Dataset.from_pandas(train_df[['text', 'label']])
test_ds = Dataset.from_pandas(test_df[['text', 'label']])

model = SetFitModel.from_pretrained(
# "BAAI/bge-m3"
# "sentence-transformers/paraphrase-multilingual-mpnet-base-v2"
"sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

args = TrainingArguments(
    batch_size=16,
    num_epochs=1,
    num_iterations=20
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    metric="accuracy"
)

trainer.train()

metrics = trainer.evaluate()
print("Evaluation Metrics:", metrics)

predictions = model.predict(test_df['text'].tolist())
print("\nClassification Report:")
print(classification_report(test_df['label'], predictions))
```
